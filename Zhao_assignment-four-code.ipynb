```python
import csv
House = {}
with open(r"C:\Users\zhaow\OneDrive\Desktop\assignment_4_property_tax_report_2019\assignment_4_property_tax_report_2019.csv",'r') as input_file:
    report = csv.reader(input_file) 
    next(report) 
    for item in report: 
        if item[24]:
            if int(item[24]) >= 1900:
                if int(item[24]) in House:
                    House[int(item[24])] +=1
                else:
                    House[int(item[24])] = 1

import matplotlib.pyplot as plt
House_sort = sorted(House.items())
x,y = zip(*House_sort)
plt.plot(x,y,color='C1')
plt.xlabel('Year')
plt.ylabel('House Built', color='C0')
plt.title('Houses built per year',color='C0')
plt.grid(color='C0')
plt.show()
```
